```python
#LinePlot

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#Customization of Line Plot

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color='red', linestyle='--', marker='o', label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
plt.legend()
plt.show()
```

```python
#Scatter plot

import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris')
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot')
plt.show()
```

```python
#Histogram

import seaborn as sns

iris = sns.load_dataset('iris')
sns.histplot(data=iris, x='petal_length')
plt.title('Histogram')
plt.show()
```

```python
#Bar Plot

import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')
sns.countplot(data=titanic, x='class')
plt.title('Bar Plot')
plt.show()
```

```python
#Pie Chart

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Box plot

import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')
sns.boxplot(data=titanic, x='class', y='age')
plt.title('Box Plot')
plt.show()
```

```python
#Violin Plot

import seaborn as sns

iris = sns.load_dataset('iris')
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.title('Violin Plot')
plt.show()
```
